

### 1. Key Features of Python
- **Simple & Readable**: Python’s syntax is clear, making it easy to learn and understand.
- **Interpreted Language**: Executes code line-by-line, making debugging easier.
- **Object-Oriented**: Supports concepts like classes and objects for structured programming.
- **Dynamic Typing**: Variable types are determined at runtime, so you don’t need to declare them.
- **Extensive Libraries**: Includes modules for web development, data science, AI, etc.
- **Cross-Platform**: Works on various operating systems without modification.

---

### 2. Role of Predefined Keywords
- **Definition**: Keywords are reserved words in Python with special meanings, like `if`, `else`, `for`, etc.
- **Example**:
  ```python
    if age >= 18:
          print("Eligible to vote")
            else:
                  print("Not eligible")

                    ```
                     

Here, `if` and `else` control the flow based on conditions.

                      

                    
  ### 3. Mutable vs Immutable Objects
    
 -                
**Mutable Objects**: Can change after creation (e.g., lists, dictionaries).        
 ```python
 my_list = [1, 2, 3]
     My_list.append(4)
 #my_list is now [1, 2, 3, 4]
              ```
 -                      
**Immutable Objects**: Cannot change once created (e.g., strings, tuples).                                
  ```python
       my_str = "Hello"
        my_str[0] = "h"
  #raises an error as strings are immutable.  
  
                         ```

### 4. Types of Operators
                            
 - **Arithmetic**: `+`, `-`, `*`, `/` (e.g., `a + b`).
 -
  **Comparison**: `==`, `!=`, `>`, `<` (e.g., `a == b`).
 -
   **Logical**: `and`, `or`, `not` (e.g., `a > b and b > c`).

 -  **Bitwise**: `&`, `|`, `^`, `~` (e.g., `a & b`).
  ```python
         a = 5
            b = 3
              print(a + b)  # Output: 8
                ```
           ---
                                     




                                               
 ### 5. Type Casting.

  - **Definition**: Converting a variable from one type to another.
  - **Example**
 ```python
                 age = "25"
                 age = int(age)
        # Converts string to integer.
        print(age + 5)  # Output: 30
         ```
